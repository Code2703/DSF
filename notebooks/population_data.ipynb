```python
# get data from oblig and manually crafted dataset regarding population changes

import pandas as pd
pop_data = pd.read_csv('../data/INFLUENZA_oblig/data.csv')
```

```python
pop_data = pop_data[['temporal', 'georegion', 'georegion_type', 'pop']]
pop_data.info()
```

```python
import pandas as pd

# Angenommen, Ihre Daten sind in einer DataFrame 'df' gespeichert und die Spalte mit den Kantonsabkürzungen heißt "Abbreviation"

# Ursprüngliches Wörterbuch der Kantonskürzel
cantons = {'Zürich':'ZH', 'Zug':'ZG', 'Uri':'UR', 'Nidwalden':'NW', 'Obwalden':'OW', 'Thurgau':'TG', 'Freiburg':'FR', 'Neuenburg':'NE', 'Genf':'GE', 'Jura':'JU', 'Wallis':'VS', 'Waadt':'VD', 'Graubünden':'GR', 'Sankt_Gallen':'SG', 'Tessin':'TI', 'Schwyz':'SZ', 'Schaffhausen':'SH', 'Luzern':'LU', 'Glarus':'GL', 'Basel-Stadt':'BS', 'Basel-Landschaft':'BL', 'Bern':'BE', 'Aargau':'AG', 'Appenzell_Innerrhoden':'AI', 'Appenzell_Ausserrhoden':'AR', 'Solothurn':'SO', 'Schweiz':'CH'}

# Wörterbuch umkehren, um die Abkürzungen als Schlüssel zu haben
cantons_reversed = {v: k for k, v in cantons.items()}

# Eine neue Spalte für die vollständigen Kantonsnamen hinzufügen
pop_data['georegion_compl'] = pop_data['georegion'].map(cantons_reversed)

# Die DataFrame anzeigen, um das Ergebnis zu überprüfen
pop_data
```

```python
cantonal_pop = pd.read_csv('../data/pop_data_cantons/data.csv')
```

```python
cantonal_pop.info()
```

```python
# Create dict for abbreviation mapping
cantons = {'Zürich':'ZH', 'Zug':'ZG', 'Uri':'UR', 'Nidwalden':'NW', 'Obwalden':'OW', 'Thurgau':'TG', 'Freiburg':'FR', 'Neuenburg':'NE', 'Genf':'GE', 'Jura':'JU', 'Wallis':'VS', 'Waadt':'VD', 'Graubünden':'GR', 'Sankt_Gallen':'SG', 'Tessin':'TI', 'Schwyz':'SZ', 'Schaffhausen':'SH', 'Luzern':'LU', 'Glarus':'GL', 'Basel-Stadt':'BS', 'Basel-Landschaft':'BL', 'Bern':'BE', 'Aargau':'AG', 'Appenzell_Innerrhoden':'AI', 'Appenzell_Ausserrhoden':'AR', 'Solothurn':'SO', 'Schweiz':'CH'}  

# Eine neue Spalte für die Abkürzungen hinzufügen
cantonal_pop['Abbreviation'] = cantonal_pop['Region'].map(cantons)
```

```python
cantonal_pop
```

```python
# Führen Sie eine Gruppierung durch und nehmen Sie nur den ersten Eintrag für jede Gruppe
cantonal_pop_grouped = cantonal_pop.groupby('Abbreviation').first().reset_index()

# Führen Sie die Daten zusammen
merged_data = pop_data.merge(cantonal_pop_grouped[['Abbreviation', 'Canton Size (km²)']], 
                             left_on='georegion', right_on='Abbreviation')

# Überprüfen Sie das Ergebnis
print(merged_data)
```

```python
# pop / canton size
merged_data['pop_density'] = merged_data['pop'] / merged_data['Canton Size (km²)']
merged_data
```

```python
import pandas as pd
from datetime import datetime

def get_weeks_in_year(year):
    """Gibt die Anzahl der Wochen in einem gegebenen Jahr zurück."""
    last_day = datetime(year, 12, 31)
    return last_day.isocalendar()[1]  # Gibt die Woche des Jahres für den letzten Tag zurück

# Angenommen, yearly_data ist Ihr DataFrame mit jährlichen Daten
# Berechnen Sie die Anzahl der Wochen für jedes Jahr
cantonal_pop['Wochen'] = cantonal_pop['Jahr'].apply(get_weeks_in_year)

# Teilen Sie die jährlichen Daten durch die Anzahl der Wochen
for col in ['Geburten', 'Todesfälle', 'Überschuss', 'Wanderungssaldo international 1']:
    cantonal_pop[col + '_woechentlich'] = cantonal_pop[col] / cantonal_pop['Wochen']

# Entfernen Sie die ursprünglichen jährlichen Spalten
cantonal_pop.drop(['Geburten', 'Todesfälle', 'Überschuss', 'Wanderungssaldo international 1'], axis=1, inplace=True)

# Führen Sie die Daten zusammen
# Angenommen, weekly_data hat auch die Spalten 'Kanton' und 'Jahr'
merged_data = merged_data.merge(cantonal_pop, on=['Abbreviation', 'Jahr'])

# Jetzt enthält merged_data sowohl die ursprünglichen wöchentlichen als auch die berechneten wöchentlichen Daten
merged_data
```
