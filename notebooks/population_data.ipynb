```python
import pandas as pd
import numpy as np

# Laden Ihrer Daten
data = pd.read_csv('../data/pop_data_cantons/final_pop_data.csv')

# Vorbereitung des DataFrames für wöchentliche Daten
weeks_per_year = 52
weekly_data = []

for _, row in data.iterrows():
    total_population = row['Bevölkerung 31. Dezember']
    
    # Berechnen der Verhältnisse für jede demografische Gruppe
    demografische_gruppen = ['Mann', 'Frau', 'Alter 0–19', 'Alter 20–64', 'Alter 65 und mehr',
                             'Schweizer', 'Ausländer ',
                             'Ledig', 'Verheiratet', 'Verwitwet', 'Geschieden', 'Unverheiratet',
                             'In eingetragener Partnerschaft', 'Aufgelöste Partnerschaft',
                             'Städtischer Kernraum', 'Einflussgebiet Stadt', 'Ausserhalb des Einflusses städtischer Kerne']
    ratios = {grp: row[grp] / total_population for grp in demografische_gruppen}

    for week in range(1, weeks_per_year + 1):
        week_data = row.to_dict()
        week_data['Woche'] = week

        # Lineare Imputation für Bevölkerungszahlen und andere Zählvariablen
        week_population = np.linspace(row['Bevölkerung 1. Januar'], row['Bevölkerung 31. Dezember'], weeks_per_year)[week - 1]

        # Anpassung der Zahlen für demografische Gruppen basierend auf dem Verhältnis
        for grp in demografische_gruppen:
            week_data[grp] = np.round(week_population * ratios[grp], 0)


        # Teilen der Werte durch 52 und Zuweisung zu jeder Woche
        for col in ['Geburten', 'Todesfälle', 'Überschuss', 'Wanderungssaldo international 1', 'Wanderungssaldo interkantonal', 'total']:
            week_data[col] = np.round(row[col] / weeks_per_year, 0)

        week_data['Bevölkerung'] = week_population

        # Einfache Kopie für BIP-bezogene Daten
        for bip_col in ['BIP in Millionen CHF', 'BIP Growth Laufende Preise', 'BIP Growth Preise Vorjahr', 'BIP pro Kopf']:
            week_data[bip_col] = row[bip_col]

        weekly_data.append(week_data)

# Konvertieren der Liste von Dictionaries in einen DataFrame
weekly_data_df = pd.DataFrame(weekly_data)

# drop Spalte Total
weekly_data_df.drop(['Total'], axis=1, inplace=True)

# Spalte % in % over Year umbenennen
weekly_data_df.rename(columns={'%':'% over Year'}, inplace=True)

# Sortieren der Daten nach Region (Kanton/Abbreviation) und Woche
weekly_data_df.sort_values(by=['Abbreviation', 'Jahr', 'Woche'], inplace=True)

# Berechnen der prozentualen Veränderung im Vergleich zur Vorwoche
weekly_data_df['% over week'] = weekly_data_df.groupby(['Abbreviation', 'Jahr'])['Bevölkerung'].pct_change() * 100

# Hinzufügen der kombinierten Jahr-Woche-Spalte
weekly_data_df['Jahr-Woche'] = weekly_data_df['Jahr'].astype(str) + '-W' + weekly_data_df['Woche'].apply(lambda x: f'{x:02d}')

# get weekly density
weekly_data_df['Pop Density'] = weekly_data_df['Bevölkerung'] / weekly_data_df['Canton Size (km²)']

# Speichern der wöchentlichen Daten
weekly_data_df.to_csv('../data/pop_data_cantons/weekly_imputed_pop_data_final.csv', index=False)
```
